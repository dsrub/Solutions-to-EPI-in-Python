```python
def BSRec(i, j, A, x):
    if i>j:
        return -1
    
    mid = (i+j)//2
    
    if A[mid] == x:
        return mid
    
    elif A[mid] < x:
        return BSRec(mid+1, j, A, x)
    
    else:
        return BSRec(i, mid-1, A, x)
    

def BS(A, x, CompFunct = lambda x:x):
    
    i, j = 0, len(A) - 1
    while i <= j:
        mid = (i+j)//2
        
        if CompFunct(A[mid]) == CompFunct(x):
            return mid
        elif CompFunct(A[mid]) < CompFunct(x):
            i = mid+1
        else: 
            j = mid - 1
        
    return -1
    

    
A = [1, 2, 3, 4, 6, 7, 8]
print(BSRec(0, len(A)-1, A, 20), BS(A, 20))
print(BSRec(0, len(A)-1, A, 5), BS(A, 5))
print(BSRec(0, len(A)-1, A, 1), BS(A, 1))
print(BSRec(0, len(A)-1, A, 4), BS(A, 4))
```

```python
### TC: O(log n), SC: O(log n) (from stack).  An O(1) SC version can be implemented iteratively
def BSleftRightRec(i, j, A, x, found_ind = -1, direction = 'Left'):
    
    if i>j:
        return found_ind

    mid = (i+j)//2

    if A[mid] == x:
        if direction == 'Left':
            return BSleftRightRec(i, mid - 1, A, x, found_ind = mid, direction = direction)
        else: 
            return BSleftRightRec(mid + 1, j, A, x, found_ind = mid, direction = direction)

    elif A[mid] < x:
        return BSleftRightRec(mid+1, j, A, x, found_ind = found_ind, direction = direction)

    else:
        return BSleftRightRec(i, mid-1, A, x, found_ind = found_ind, direction = direction)

def BSLR(A, x, direction = 'Left'):
    return BSleftRightRec(0, len(A)-1, A, x, found_ind = -1, direction = direction)


A = [1, 1, 2, 2, 2, 2, 3, 4,4,4, 5, 6, 6,7, 8, 9, 10]
print(BSLR(A, 1, direction = 'Left'))
print(BSLR(A, 2, direction = 'Left'))
print(BSLR(A, 4, direction = 'Left'))
print(BSLR(A, 8, direction = 'Left'))
print(BSLR(A, 15, direction = 'Left'))

print(BSLR(A, 1, direction = 'Right'))
print(BSLR(A, 2, direction = 'Right'))
print(BSLR(A, 4, direction = 'Right'))
print(BSLR(A, 8, direction = 'Right'))
print(BSLR(A, 15, direction = 'Right'))
```

```python
### algo to find first occurence of value greater than a specified value:


### TC: O(log n), SC: O(log n) (from stack).  An O(1) SC version can be implemented iteratively
def BSGreaterRec(i, j, A, x, found_ind):
    
    if i>j:
        if found_ind == -1:
            return 'last element in array is x'
        else:
            return A[found_ind]
    

    mid = (i+j)//2
                
    if A[mid] <= x:
        return BSGreaterRec(mid + 1, j, A, x, found_ind)
                
    else:
        return BSGreaterRec(i, mid-1, A, x, mid)
                

def BSGreater(A, x):
    return BSGreaterRec(0, len(A)-1, A, x, -1)


A = [1, 1, 2, 2, 2, 2, 4, 4, 4, 5, 6, 6,7, 8, 9, 10]
print(BSGreater(A, 1))
print(BSGreater(A, 2))
print(BSGreater(A, 3))
print(BSGreater(A, 4))
print(BSGreater(A, 5))
print(BSGreater(A, 6))
print(BSGreater(A, 7))
print(BSGreater(A, 8))
print(BSGreater(A, 9))
print(BSGreater(A, 10))
print(BSGreater(A, 11))
```

```python
#### TC: O(log n)
    
def Enclosed(A, x):
    first = BSleftRightRec(0, len(A)-1, A, x, found_ind = -1, direction = 'Left')
    
    if first == -1:
        return (-1, -1)
    else:
        last = BSleftRightRec(0, len(A)-1, A, x, found_ind = -1, direction = 'Right')
    
    return (first, last)

A = [1, 1, 2, 2, 2, 2, 3, 4,4,4, 5, 6, 6,7, 8, 9, 10]
print(Enclosed(A, 4))
print(Enclosed(A, 12))
```

```python
### this is a simple extension of BS, where I for an input string s, I compare the 1st |s| of the mid element
### in each iteration of BS and go the appropriate direction.  TC: O(|S|log n), SC:O(|S|log n) (which can be brought 
### iteratively down to O(1))

def BSString(i, j, A, s):
    
    if i>j:
        return -1
    
    mid = (i+j)//2
    
    
    if len(A[mid]) >= len(s):

        if A[mid][0:len(s)] == s:
            return mid

        elif A[mid][0:len(s)] < s:
            return BSString(mid+1, j, A, s)

        else:
            return BSString(i, mid-1, A, s)
    
    
    else:
        if A[mid] < s:
            return BSString(mid+1, j, A, s)

        else:
            return BSString(i, mid-1, A, s)
    

A = ['this', 'is', 'my', 'amazing', 'sample', 'text', 'string', 'to', 'be', 'searched']
A.sort()
print(A)
print(BSString(0, len(A)-1, A, 'am'))
print(BSString(0, len(A)-1, A, 'tex'))
print(BSString(0, len(A)-1, A, 'boo'))
```

```python
def helper(i, j, A):
    if i > j:
        return -1
    
    mid = (i+j)//2
    
    if A[mid] == mid:
        return mid
    elif A[mid] < mid:
        return helper(mid+1, j, A)
    else:
        return helper(i, mid-1, A)
    
    
def SearchVal_Equals_Ind(A):
    return helper(0, len(A)-1, A)


print(SearchVal_Equals_Ind([-6, -5, -4, 3, 4, 5, 8, 9]))

print(SearchVal_Equals_Ind([-6, -5, -4, -3, -2, -1, 0, 1]))
```

```python
### integer SQRT.
### BF would be to go 1 by 1 starting from 0 up until ceil(SQRT(n)), which would take O(sqrt(n)) time.
### we can do better with a BS

def IntSqrt(n):
    best = 0
    i, j= 0, n
    
    while i <= j:
        mid = (i+j)//2
        if mid**2 == n:
            return mid
        elif mid**2 < n:
            best = mid
            i = mid + 1
        else:
            j = mid - 1
            
    return best
           
print(IntSqrt(0))
print(IntSqrt(1))
print(IntSqrt(2))
print(IntSqrt(3))
print(IntSqrt(4))
print(IntSqrt(5))
print(IntSqrt(6))
print(IntSqrt(7))
print(IntSqrt(99))
print(IntSqrt(100))
```

```python
### I can decompose the search range into 3 specific intervals depending on the value of x
### A simple analysis shows that TC: O(log(x/d)), where d is a prespecified tolerance
def ComputeRealSQRT(x, d):
    if x == 0 or x == 1 or x == 4:
        return (0, 1, None, None, 2)[x]
    elif 0 < x < 1:
        lower, upper = x, 1
    elif 1 < x < 4:
        lower, upper = 1, x
    else:
        lower, upper = 1, x/2.
    
    while True:
    
        mid = (upper+lower)/2
        
        if ((mid-d)**2 <= x) and (x<= (mid+d)**2):
            return mid
        elif x < (mid-d)**2:
            upper = mid  
        else:
            lower = mid
    

print(ComputeRealSQRT(1, .001))
print(ComputeRealSQRT(.1, .001))
print(ComputeRealSQRT(2, .001))
print(ComputeRealSQRT(100, .00001))
print(ComputeRealSQRT(24, .001))
print(ComputeRealSQRT(36, .001))
print(ComputeRealSQRT(212341244, .0001))
```

```python
## the following algo does a binary search like search to find x/y given a tolerance d.  
## A simple analysis shows that if y>x the TC is O(log(1/d)) and if x>y, the TC is O(log(x/d)). 
### In both cases the space complexity is O(1)
def ComputeDivision(x, y, d, max_iter = 1000):
    
    if x == y:
        return 1
    elif y>x:
        lower, upper = 0, 1
    else:
        lower, upper = 1, x
        
    iters = 1
    
    while iters < max_iter:
        iters += 1

        mid = (upper+lower)/2

        if y*(mid+(-1*d)) <= x <= y*(mid+d):
            return mid
        elif x < y*(mid+(-1*d)):
            upper = mid  
        else:
            lower = mid
            
    
    return "Did not converge"

            
print(ComputeDivision(20, 5, 1e-10))  #4
print(ComputeDivision(5, 20, 1e-10)) # .25
print(ComputeDivision(1234, 33, 1e-10)) ## ~37.39
print(ComputeDivision(23123, 29312348420394802394873240987, 1e-28)) ##7.8884843e-25
print(ComputeDivision(5, .3, .001)) ##7.8884843e-25
```

```python
### TC: O(m+n), SC: O(m+n)
def SearchMatrixRec(i, j, M, val):
    
    if i == len(M) or j == -1:
        return False
    
    elif M[i][j] == val:
        return True
    
    elif M[i][j] < val:
        return SearchMatrixRec(i+1, j, M, val)
    
    else:
        return SearchMatrixRec(i, j-1, M, val)
    
### TC: O(m+n), SC: O(1)
def SearchMatrix(M, val):
    
    i, j = 0, len(M[0])-1
    
    while i <= len(M)-1 and j >= 0:

        if M[i][j] == val:
            return True
        
        elif M[i][j] < val:
            i += 1
            
        else:
            j -= 1
    
    return False

M = [[-1, 2, 4, 4, 6], 
     [1, 5, 5, 9, 21], 
     [3, 6, 6, 9, 22], 
     [3, 6, 8, 10, 24], 
     [6, 8, 9, 12, 25], 
     [8, 10, 12, 13, 40]]

print(SearchMatrixRec(0, len(M[0])-1, M, 6))
print(SearchMatrix(M, 6))

print(SearchMatrixRec(0, len(M[0])-1, M, -1))
print(SearchMatrix(M, -1))

print(SearchMatrixRec(0, len(M[0])-1, M, 40))
print(SearchMatrix(M, 40))

print(SearchMatrixRec(0, len(M[0])-1, M, 100))
print(SearchMatrix(M, 100))

print(SearchMatrixRec(0, len(M[0])-1, M, -40))
print(SearchMatrix(M, -40))
```

```python
from random import randint, shuffle

## Expected TC: O(n), Expected SC: O(log n)

def RandomPart(l, r, A):
    R = randint(l, r)
    p = A[R]
    A[l], A[R] = A[R], A[l]
    
    i, j = l+1, l + 1 
    
    while j <= r:
        if A[j] > p:
            j += 1
        else:
            A[i], A[j] = A[j], A[i]
            i += 1; j += 1
            
    A[l], A[i-1] = A[i-1], A[l]
    
    return i-1
        

def SmallestK(l, r, k, A):
    if l == r:
        return A[l]
    
    Q = RandomPart(l, r, A)
    
    if Q == l+k-1:
        return A[Q]
    elif Q < l+k-1:
        return SmallestK(Q+1, r, k-Q+l-1, A)
    else:
        return SmallestK(l, Q-1, k, A)
    
def KthLargest(A, k):
    k = len(A) - k + 1
    return SmallestK(0, len(A)-1, k, A)
    
A = [2, 5, 3, 7, 1, 4, 6]
shuffle(A)
print(KthLargest(A, 1))
print(KthLargest(A, 2))
print(KthLargest(A, 3))
print(KthLargest(A, 4))

print("----------")

A = [1, 2, 2, 3, 3, 3, 4, 5, 6]
shuffle(A)
print(KthLargest(A, 1))
print(KthLargest(A, 2))
print(KthLargest(A, 3))
print(KthLargest(A, 4))
print(KthLargest(A, 5))
print(KthLargest(A, 6))
```

```python
#TC: O(n), exactly 3/2n-2 comparisons, SC: O(1)
def FindMinMax(A):

    if A[0] > A[1]:
        curr_max = A[0]
        curr_min = A[1]
    else:
        curr_max = A[1]
        curr_min = A[0]


    if len(A) % 2 == 0:
        upper = len(A)
    else:
        upper = len(A) - 1

    for i in range(2, upper, 2):
        if A[i] > A[i+1]:
            curr_max = max(curr_max, A[i])
            curr_min = min(curr_min, A[i+1])
        else:
            curr_max = max(curr_max, A[i+1])
            curr_min = min(curr_min, A[i])

    if len(A) % 2 != 0:
        curr_max = max(curr_max, A[-1])
        curr_min = min(curr_min, A[-1])

    return (curr_min, curr_max)
    
    
A = [8, 15, 6, 22, 32, 5, 4, 12, 18, 14, 2]
print(FindMinMax(A))

A = [8, 15, 6, 22, 32, 5, 4, 12, 14, 2]
print(FindMinMax(A))
```

```python
#TC: O(n), SC: O(n)
def FindMissingDup1(A):
    H={}
    n = len(A)
    
    for a in A:
        if a in H:
            dup = a
        else:
            H[a] = 1
            
    miss = int(n*(n-1)/2+dup-sum(A))
    
    return (dup, miss)

```
